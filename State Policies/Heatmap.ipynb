```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set(style = "whitegrid", 
        color_codes = True,
        font_scale = 1.5)
sns.set_palette('Blues') 
```

```python
SB = pd.read_csv('Small Business Pulse Survey.csv')
Relaxing_Policy = pd.read_csv('relaxing-policies.csv')
Restrictive_Policy = pd.read_csv('restrictive-policies.csv')
```

```python
SB.head()
```

```python
Relaxing_Policy.head()
```

```python
Restrictive_Policy.head()
```

```python
relax = SB.merge(Relaxing_Policy, left_on='Full Name', right_on='State', how='outer')
relax.head()
```

```python
restrict = SB.merge(Restrictive_Policy, left_on='Full Name', right_on='State', how='outer')
restrict.head()
```

```python
test = restrict[['Negative Impact', 'Stay at home/ shelter in place','Closed non-essential businesses',
                   'Closed restaurants except take out','Closed gyms','Closed movie theaters',
                   'Total # Restrictive Policies']]
test
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(test.corr(), annot=True, cmap="Oranges");
sns.set(font_scale=1)
```

```python
test1 = relax[['Negative Impact', 'End/relax stay at home/shelter in place','Began to reopen businesses',
               'Reopen restaurants','Reopened gyms','Reopened movie theaters','Total # Relaxing Policies']]
test1
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(test1.corr(), annot=True, cmap="BuGn");
sns.set(font_scale=1)
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(restrict.corr(), annot=True, cmap="BuGn");
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(relax.corr(), annot=True, cmap="BuGn");
```
